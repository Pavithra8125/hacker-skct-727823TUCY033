```python
import random

ips = ['192.168.1.5', '10.0.0.9', '172.16.0.3']
users = ['root', 'admin', 'pavithra']

f = open('/tmp/auth_test.log', 'w')
for i in range(50):
    ip = random.choice(ips)
    user = random.choice(users)
    f.write('Mar 29 12:00:0{} webserver sshd[1234]: Failed password for {} from {} port 22 ssh2\n'.format(i%9, user, ip))
f.close()

print("Log file created with 50 fake entries!")
```

```python
# Test Case 2: Multiple Users Attacked from Same IP
users_per_ip = {}
f = open('/tmp/auth_test.log', 'r')
for line in f:
    if 'Failed' in line:
        ip = line.split()[10]
        user = line.split()[8]
        if ip not in users_per_ip:
            users_per_ip[ip] = []
        users_per_ip[ip].append(user)
f.close()

print("=== TC2: Multiple Users Attacked ===")
for ip, users in users_per_ip.items():
    if len(users) > 3:
        print("ALERT: IP", ip, "tried usernames:", set(users))
```

```python
# Test Case 3: Root Login Attempts
spe = {}
f = open('/tmp/auth_test.log', 'r')
for line in f:
    if 'Failed' in line and 'root' in line:
        ip = line.split()[10]
        spe[ip] = spe.get(ip, 0) + 1
f.close()

print("=== TC3: Root Login Attempts ===")
for ip, count in spe.items():
    if count > 3:
        print("ALERT: Root attack from", ip, "| Attempts:", count)
```
